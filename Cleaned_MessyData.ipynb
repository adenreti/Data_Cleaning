# Cleaning messy_data.csv

## Importing our library and dataset

```python
import pandas as pd
```

```python
df=pd.read_csv('messy_data.csv', encoding='Latin-1', sep=';')
```

```python
df
```

## Dropping Column: Unnamed: 8

```python
df=df.drop('Unnamed: 8', axis=1)
```

```python
df
```

## Changing column titles to lowercase

```python
df.columns = df.columns.str.lower()
```

```python
df
```

## Renaming column titles with special characters and space 

```python
df=df.rename(columns={'imbd title id': 'imbd_title_id', 'original titlê': 'movie_title', 'release year': 'release_year',
                  'genrë¨': 'genre', 'content rating': 'content_rating'})
```

```python
df
```

## Getting dataset info

```python
df.info()
```

## Changing the datatypes

```python
df.score = pd.to_numeric((df.score), errors='coerce')
```

```python
df
```

## Filling null values in content_rating column

```python
df.content_rating.fillna('Unrated')
```

```python
df.content_rating=df.content_rating.fillna('Unrated')
```

```python
df
```

## Cleaning votes column

```python
df.rename(columns={' votes ': 'votes'}, inplace=True)
```

```python
df.votes=df.votes.str.replace('.', '', regex=False)
```

```python
df=df.drop([13])
```

```python
df.votes=df.votes.astype(int)
```

```python
df
```

## Converting release_year column to datetime data type

```python
df.release_year = pd.to_datetime(df.release_year, errors='coerce', infer_datetime_format=True)
```

```python
df
```

## Extracting the first genre on the genre column

```python
df.genre = df.genre.str.split(',', n=1, expand=True)[0]
```

```python
df
```

```python
df.country.replace({'US':'USA', 'US.':'USA', 'New Zesland':'New Zealand', 'New Zeland':'New Zealand', 
                    'Italyl':'Italy', 'West Germany':'Germany'}, inplace=True)
```

```python
df
```

```python
df.drop(['genre1', 'genre2'], axis=1, inplace=True)
```
